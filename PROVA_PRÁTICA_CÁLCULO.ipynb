```python

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo a variável simbólica x
x = sp.symbols('x')

# Definindo a função f(x)
f = 3*x**2 + 2*x**2 - x + 1

# Calculando o limite
limite = sp.limit(f, x, -2)

# Imprimindo o valor do limite
print("Limite:", limite)

# Convertendo a função sympy para uma função numpy
f_np = sp.lambdify(x, f, 'numpy')

# Criando um intervalo de valores de x
x_values = np.linspace(-3, 1, 400)

# Calculando os valores correspondentes de f(x)
y_values = f_np(x_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = 3x^2 + 2x^2 - x + 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de $f(x)$')
plt.axvline(x=-2, color='r', linestyle='--', label='x = -2')
plt.axhline(y=limite, color='g', linestyle='--', label=f'Limite = {limite}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo a variável simbólica x
x = sp.symbols('x')

# Definindo a função f(x)
f = (x**2 - 9) / (x**2 + x - 12)

# Calculando o limite
limite = sp.limit(f, x, 3)

# Imprimindo o valor do limite
print("Limite:", limite)

# Convertendo a função sympy para uma função numpy
f_np = sp.lambdify(x, f, 'numpy')

# Criando um intervalo de valores de x
x_values = np.linspace(2.5, 3.5, 400)

# Calculando os valores correspondentes de f(x)
y_values = f_np(x_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = \\frac{x^2 - 9}{x^2 + x - 12}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de $f(x)$')
plt.axvline(x=3, color='r', linestyle='--', label='x = 3')
plt.axhline(y=limite, color='g', linestyle='--', label=f'Limite = {limite}')
plt.ylim(-10, 10)  # Ajustando o intervalo y para melhor visualização
plt.legend()
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Definindo a função para o limite
def f(x):
    return (np.sin(x) - np.sin(np.pi)) / (x - np.pi)




# Definindo o valor conhecido do limite
limite = -1
print("limite:" , limite)

# Criando um intervalo de valores de x próximos a pi
x_values = np.linspace(np.pi - 0.1, np.pi + 0.1, 400)

# Calculando os valores correspondentes de f(x)
y_values = f(x_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = \\frac{\\sin(x) - \\sin(\\pi)}{x - \\pi}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de $f(x)$')
plt.axvline(x=np.pi, color='r', linestyle='--', label='$x = \pi$')
plt.axhline(y=limite, color='g', linestyle='--', label=f'Limite = {limite}')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    if x <= 1:
        return 2*x - 1
    elif x <= 2:
        return x + 1
    else:
        return (x**2 + 2*x - 8) / (2*x - 4)

# Gerando valores de x próximos a 1 e 2
x_values = np.linspace(0.5, 2.5, 400)
y_values = [f(x) for x in x_values]

# Plotando o gráfico da função
plt.plot(x_values, y_values, label='$f(x)$')

# Calculando os limites laterais para x = 1
limite_esquerda_1 = 2*1 - 1  # Limite à esquerda de x = 1
limite_direita_1 = 1 + 1     # Limite à direita de x = 1
plt.scatter([1], [limite_esquerda_1], color='red', label='Limite esquerda em $x = 1$')
plt.scatter([1], [limite_direita_1], color='blue', label='Limite direita em $x = 1$')

# Calculando os limites laterais para x = 2
limite_esquerda_2 = 2 + 1  # Limite à esquerda de x = 2
limite_direita_2 = np.nan  # Definindo como NaN temporariamente
for i in range(400):
    if x_values[i] > 2:
        limite_direita_2 = y_values[i]
        break

plt.scatter([2], [limite_esquerda_2], color='green', label='Limite esquerda em $x = 2$')
plt.scatter([2], [limite_direita_2], color='purple', label='Limite direita em $x = 2$')

# Adicionando legendas e rótulos
plt.title('Gráfico da função e Limites laterais')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return (3*x + 5) / (x - 4)

# Definindo o intervalo de valores de x
x_values = np.linspace(1, 20, 400)  # Usando um intervalo maior para evitar divisão por zero

# Calculando os valores correspondentes de f(x)
y_values = f(x_values)

# Calculando o limite quando x se aproxima do infinito
limite = 3

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = \\frac{3x + 5}{x - 4}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de $f(x)$')
plt.axhline(y=limite, color='g', linestyle='--', label=f'Limite = {limite:.2f}')

plt.axvline(x=x_values[-1], color='r', linestyle='--', label=f'x = {x_values[-1]:.2f}')

plt.legend()
plt.grid(True)
plt.show()

# Indicando o valor do limite no terminal
print(f"O valor do limite é: {limite}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função para o limite
def f(x):
    return np.sqrt((12*x**3 - 5*x + 2) / (1 + 4*x**2 + 3*x**3))

# Definindo o intervalo de valores de x
x_values = np.linspace(1, 10, 400)  # Usando um intervalo maior para evitar divisão por zero

# Calculando os valores correspondentes de f(x)
y_values = f(x_values)

# Calculando o limite quando x se aproxima do infinito
limite = f(1e10)

# Imprimindo o valor do limite
print("Limite quando x se aproxima do infinito:", limite)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = \sqrt{\\frac{12x^3 - 5x + 2}{1 + 4x^2 + 3x^3}}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de $f(x)$')
plt.axhline(y=limite, color='g', linestyle='--', label=f'Limite = {limite:.2f}')
plt.axvline(x=x_values[-1], color='r', linestyle='--', label=f'x = {x_values[-1]:.2f}')

plt.legend()
plt.grid(True)
plt.show()
```
